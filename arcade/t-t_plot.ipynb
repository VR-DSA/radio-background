# Derive spectral index map from 408 MHz & 1.4 GHz data

```python
import matplotlib.pyplot as plt
import pylab
import numpy as np
import healpy as hp
from scipy import stats
from scipy.optimize import curve_fit
import sys
sys.path.append('../spherical_harms')
import ModelDefinitions as MD
import LogProb as LP
from const import *
from IPython.display import display, Math
%matplotlib inline

# constants besides in const.py
NSIDE_a = 32
rad = 5.*np.pi/180.
```

```python
# functions to generate and analyse t-t plot for specific location

def func(x,beta,c):
    return x*(1.4/0.408)**(-beta) + c

def tt_plot(theta,phi,m1,m2,ns=NSIDE_a,return_values=False):
    
    # get data
    p1 = hp.ang2vec(theta,phi)
    x_p1 = m1[hp.query_disc(ns,p1,rad)]
    y_p1 = m2[hp.query_disc(ns,p1,rad)]
    
    # fit line
    popt,pcov = curve_fit(func,x_p1,y_p1,p0=[2.5,2.])
    
    # return
    if return_values:
        return popt,pcov,x_p1,y_p1
    return popt,pcov
```

Two maps are considered: the 408 MHz Haslam map (with no filtering) and the 1.4 GHz map. These data are to be used to derive a spectral index map of the sky, with associated error, using the t-t plot technique. 

The maps are first read in and degraded to NSIDE=32. 5-deg diameter regions are then used to derive local t-t plots, which are fit with straight lines. 

```python
# read maps
m408 = hp.read_map('data/lambda_haslam408_nofilt.fits')/1000.
m14 = hp.read_map('data/STOCKERT+VILLA-ELISA_1420MHz_1_256.fits')/1000.

# lower resolution
m408_dg = hp.pixelfunc.ud_grade(m408, NSIDE_a)
m14_dg = hp.pixelfunc.ud_grade(m14, NSIDE_a)

# example t-t plots for two locations, together with line fits

# from TRIS: 10h and 20.5h and dec +42
#thet1 = 0.66
#phi1 = 3.107
thet1 = 0.43723841646613837
phi1 = 0.04619989196455578
thet2 = 1.541
phi2 = 1.404

z,cov,x_p1,y_p1 = tt_plot(thet1,phi1,m408_dg,m14_dg,return_values=True)

print 'HALO LOCATION'
print z[0],z[1]
print np.sqrt(cov[0,0]),np.sqrt(cov[1,1])

plt.figure(figsize=(10.,7.))
plt.plot(x_p1,y_p1,'o',label='p1')
plt.plot(x_p1,func(x_p1,*z),'-',label='fit')
plt.xlabel('T_408 (K)')
plt.ylabel('T_1400 (K)')
plt.legend()

z,cov,x_p1,y_p1 = tt_plot(thet2,phi2,m408_dg,m14_dg,return_values=True)

print 'PLANE LOCATION'
print z[0],z[1]
print np.sqrt(cov[0,0]),np.sqrt(cov[1,1])

plt.figure(figsize=(10.,7.))
plt.plot(x_p1,y_p1,'o',label='p1')
plt.plot(x_p1,func(x_p1,*z),'-',label='fit')
plt.xlabel('T_408 (K)')
plt.ylabel('T_1400 (K)')
plt.legend()
```

We now derive a map of the T spectral index and uncertainty in every pixel in the sky, and display it.

```python
# get spectral indices
print 'Getting spectral indices...'
spindex_map = m14_dg.copy()
spindex_err = m14_dg.copy()
idxs = np.arange(len(spindex_map))
for i in idxs:
    theta,phi = hp.pix2ang(NSIDE_a,i)
    z,cov = tt_plot(theta,phi,m408_dg,m14_dg,return_values=False)
    spindex_map[i] = z[0]
    spindex_err[i] = np.sqrt(cov[0,0])
    
print 'Done'
```

```python
# display maps
smap = spindex_map.copy()
serr = spindex_err.copy()
smap[spindex_err>1.] = None
serr[spindex_err>1.] = None

hp.mollview(smap,title='Temperature spectral index',unit='alpha')
hp.mollview(serr,title='Spectral index error',unit='err',max=0.1)
```

We now derive T_gal

```python
# get all spectral indices and temperatures where the spectral index error is below some threshold
err_thresh = 3.*np.nanmin(spindex_err)
idxs = np.where(spindex_err<err_thresh)[0]
spindexs = spindex_map[idxs]
spindexes = spindex_err[idxs]
T408 = m408_dg[idxs]
T14 = m14_dg[idxs]
sidxs = np.argsort(spindexs)
T14 = T14[sidxs]
T408 = T408[sidxs]
spindexes = spindexes[sidxs]
spindexs = spindexs[sidxs]

# print properties of min and max spindexes
i1 = np.where(spindex_map==np.min(spindexs))[0][0]
i2 = np.where(spindex_map==np.max(spindexs))[0][0]

print 'Minimum spectral index...'
print hp.pix2ang(NSIDE_a,i1)
print spindexs[0],spindexes[0]
print T408[0],T14[0]
print

print 'Maximum spectral index...'
print hp.pix2ang(NSIDE_a,i2)
print spindexs[-1],spindexes[-1]
print T408[-1],T14[-1]
print


# derive estimates of T_gal based on eq 5 of https://iopscience.iop.org/article/10.1086/592134/pdf 

m0 = (0.408/1.4)**(-spindexs[0])
m0_e = spindexes[0]*np.log(0.408/1.4)

m = (1.4/0.408)**(-spindexs[-1])
m_e = spindexes[-1]*np.log(0.408/1.4)

a = T408[0] - T408[-1]
b = T14[0] - T14[-1]
est_Tgal = (a - b*m)/(m0-m)
est_Tgal_e = np.sqrt(((a-b*m0)/(m0-m)**2.)**2.*m_e**2. + ((b*m-a)/(m0-m)**2.)**2.*m0_e**2.)


print est_Tgal,est_Tgal_e
    
    
```

```python
m14_gal = m14_dg.copy()
T_orig = m14_gal[i1]
for i in range(len(m14_gal)):
    m14_gal[i] = (m14_gal[i] - T_orig) + est_Tgal

m14_gal[m14_gal<0.0] = None
hp.mollview(m14_gal,title='T_gal_1400',unit='K',max=5.)

```
